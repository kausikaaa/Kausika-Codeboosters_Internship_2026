```python
import pandas as pd
print("Libraries imported successfully")
```

```python
df=pd.read_csv('student_performance.csv')
print('Dataset loaded successfully')
```

```python
print("Shape of dataset:" , df.shape)
print("Number of students:",df.shape[0])
print("Number of columns:",df.shape[1])
```

```python
df.describe()
```

```python
print("Missing values in each column:")
print(df.isnull().sum())
```

```python
print("Total missing values:")
print(df.isnull().sum().sum())
```

```python
print("Student Names:")
print(df['name'])
```

```python
print("Departments in this dataset:")
print(df['department'].unique())

print("\nHow many students per department")
print(df['department'].value_counts())
```

```python
cs_students=df[df['department']=='Computer Science']

print(f"Number of Computer Science Students: {len(cs_students)}")
```

```python
high_attendance =df[df['attendance_percentage']>90]
print(f"Number of students with high attendance:{len(high_attendance)}")
print(high_attendance[['name','department','attendance_percentage']])
```

```python
df['total_score']=df['math_score']+df['science_score']+df['programming_score']+df['english_score']

top_student=df.loc[df['total_score'].idxmax()]

print("Top student : ")
print(f"Name : {top_student['name']}")
print(f"Department : {top_student['department']}")
print(f"Total Score : {top_student['total_score']}")
print(f"Attendance : {top_student['attendance_percentage']}")
```

```python
#calculate average score per department
dept_avg =df.groupby('department')['total_score'].mean().round(2)

print("Average total score byy department:")
print(dept_avg)
```

```python
df_sorted=df.sort_values(by='total_score',ascending=False)
print("top 10 student:")
print(df_sorted[['name','department','total_score','attendance_percentage']].head(10))
```
