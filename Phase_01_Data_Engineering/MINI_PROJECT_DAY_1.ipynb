```python
import pandas as pd
print("=" * 65)
print("        STUDENT PERFORMANCE EXPLORER USING PANDAS")
print("=" * 65)
df=pd.read_csv('student_performance.csv')
print("Datas are loaded")

df["total_score"] = df["math_score"] + df["programming_score"]
print("\nFIRST 5 RECORDS")
print("-" * 65)

print(df.head())
print("\nDATASET OVERVIEW")
print("-" * 65)

total_students = df.shape[0]
total_columns = df.shape[1]

print(f"Total Number of Students : {total_students}")
print(f"Total Number of Columns  : {total_columns}")
print("\nDEPARTMENT-WISE ANALYSIS")
print("-" * 65)

department_count = df["department"].value_counts()

for department, count in department_count.items():
    print(f"{department} : {count} students")
    print("\nSCORE ANALYSIS")
print("-" * 65)

highest_math_score = df["math_score"].max()
lowest_math_score = df["math_score"].min()

average_math_score = df["math_score"].mean()
average_programming_score = df["programming_score"].mean()
average_attendance = df["attendance_percentage"].mean()

print(f"Highest Math Score              : {highest_math_score}")
print(f"Lowest Math Score               : {lowest_math_score}")
print(f"Average Math Score              : {average_math_score:.2f}")
print(f"Average Programming Score       : {average_programming_score:.2f}")
print(f"Average Attendance Percentage   : {average_attendance:.2f}%")
print("-" * 65)

gender_average = df.groupby("gender")["total_score"].mean()

for gender, average_score in gender_average.items():
    print(f"{gender} Average Total Score : {average_score:.2f}")
    print("\nTOP 5 PERFORMERS")
print("-" * 65)

top_students = df.sort_values(by="total_score", ascending=False).head(5)

for index, row in top_students.iterrows():

    print(f"Student Name : {row['name']}")
    print(f"Department   : {row['department']}")
    print(f"Total Score  : {row['total_score']}")
    print("-" * 40)
    print("\nATTENDANCE ANALYSIS")
print("-" * 65)

low_attendance_students = df[df["attendance_percentage"] < 75]

if low_attendance_students.empty:

    print("All students have attendance above 75%")

else:

    for index, row in low_attendance_students.iterrows():

        print(f"Name                  : {row['name']}")
        print(f"Department            : {row['department']}")
        print(f"Attendance Percentage : {row['attendance_percentage']}%")
        print("-" * 40)
        print("\nProject Completed Successfully")
print("=" * 65)
```
